```python
import os
from collections import defaultdict
generated_data = os.listdir("generated_data_1")
```

```python
dataset = defaultdict(dict)
# template = {"nl": [], "bash": []}
prev_filelength = None
for i, file in enumerate(generated_data):
    with open(os.path.join("generated_data_1", file), "r") as f:
        content = f.read()
        filelength = len(content.split("\n"))
        if filelength != 100:
            print(f"Warning: File {file} contains {filelength} lines")
    if i % 2 == 0:
        dataset["bash"][file] = content
    else:
        dataset["nl"][file] = content

invalid_files = []

bash = []
nl = []

for fname in dataset["bash"]:
    bash_content = dataset["bash"][fname]
    nl_content = dataset["nl"][fname.replace("cm", "nl")]
    bash_lines = bash_content.split("\n")
    nl_lines = nl_content.split("\n")
    if len(bash_lines) != len(nl_lines):
        print(f"Warning: File {fname} has {len(bash_lines)} bash lines but {len(nl_lines)} nl lines")
        invalid_files.append(fname)
    else:
        bash += bash_lines
        nl += nl_lines

print(f"invalid files count: {len(invalid_files)}")
```

```python
with open("generated_400_1.cm", "w") as f:
    f.write("\n".join(bash))
with open("generated_400_1.nl", "w") as f:
    f.write("\n".join(nl))
```

```python
import sys
module_path = os.path.abspath(os.path.join('clai/submission-code/src/'))
# print(module_path)
sys.path.insert(0, module_path)

from clai.utils.bashlint.data_tools import ast2template, bash_parser

generated_dataset = {"bash": [], "nl": []}

for n, b in zip(nl, bash):
    try:
        generated_dataset["bash"].append(ast2template(bash_parser(b)))
        generated_dataset["nl"].append(n)
    except Exception as e:
        #print(f"Error processing bash line: {bash}")
        #print(f"Error: {e}")
        pass
    
with open("dataset_postprocessed/generated_400_1.cm", "w") as f:
    f.write("\n".join(generated_dataset["bash"]))

with open("dataset_postprocessed/generated_400_1.nl", "w") as f:
    f.write("\n".join(generated_dataset["nl"]))

print(f"Post-processed dataset contains {len(generated_dataset['bash'])} bash lines \nPre-processed contains {len(bash)} bash lines")
```
